#Görev 3: Çıktıyı PRICE’a göre sıralayınız.

```python
agg_df = data.groupby(["COUNTRY", "SOURCE","SEX","AGE"]).agg({"PRICE":"mean"}).sort_values("PRICE", ascending = False)
```

```python
agg_df
```

#Görev 4: Indekste yer alan isimleri değişken ismine çeviriniz.


```python
agg_df = agg_df.reset_index()
```

```python
agg_df
```

#Görev 5: Age değişkenini kategorik değişkene çeviriniz ve agg_df’e ekleyiniz.

```python
araliklar = [0, 18, 23, 30, 40, 70]
agg_df['AGE_CAT'] = pd.cut(agg_df['AGE'], bins=araliklar, labels=["0_18", "19_23", "24_30", "31_40", "41_70"])
```

```python
agg_df
```

#Görev 6: Yeni seviye tabanlı müşterileri (persona) tanımlayınız.

```python
cat_cols = ["COUNTRY","SOURCE","SEX","AGE_CAT"]
```

```python
agg_df[cat_cols]
```

```python
agg_df["customers_level_based"] = [value_list[0].upper()+"_"+value_list[1].upper()+"_"+value_list[2].upper()+"_"+value_list[3] for value_list in agg_df[cat_cols].values]
```

```python
agg_df["PRICE"] = agg_df.groupby("customers_level_based")["PRICE"].transform("mean")
```

```python
agg_df.head()
```

#Görev 7: Yeni müşterileri (personaları) segmentlere ayırınız.

```python
segment = pd.qcut(agg_df["PRICE"], 4, labels=["D","C","B","A"])
```

```python
segment
```

```python
agg_df["SEGMENT"] = segment
```

```python
agg_df
```

```python
agg_df.groupby("SEGMENT").agg({"PRICE":["mean","max","sum"]})
```
